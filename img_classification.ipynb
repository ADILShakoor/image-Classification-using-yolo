```python
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
```

```python
class ModelTrainer:
    """Handles the training of YOLO models."""
    def __init__(self, model_architecture="yolov8m.pt"):
        self.model_architecture = model_architecture
        self.model = YOLO(model_architecture)

    def train_model(self, data_path, epochs=20, img_size=640, save_dir="runs/detect/train"):
        print("Starting model training...")
        self.model.train(data=data_path, epochs=epochs, imgsz=img_size)
        print(f"Model training complete. Results saved in {save_dir}")
```

```python
class YOLOModel:
    """Handles model loading and inference."""
    def __init__(self, model_path="runs/detect/train/weights/best.pt"):
        self.model = YOLO(model_path) if model_path else None

    def load_model(self, model_path):
        print(f"Loading model from {model_path}...")
        self.model = YOLO(model_path)

    def predict(self, image_path, save=True, save_txt=True, img_size=640, conf=0.2):
        print(f"Running prediction on {image_path}...")
        return self.model.predict(image_path, save=save, save_txt=save_txt, imgsz=img_size, conf=conf)
```

```python
class ImageProcessor:
    """Handles image visualization."""
    @staticmethod
    def display_images(image_directory, num_rows=4, num_cols=4):
        # List all image files in the directory
        image_files = [f for f in os.listdir(image_directory) if f.endswith(('.jpg', '.jpeg'))]
        print(f"Found {len(image_files)} images in {image_directory}")

        # Set up the plot
        fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))
        
        # Display images in a grid
        for i, image_file in enumerate(image_files[:num_rows * num_cols]):
            image_path = os.path.join(image_directory, image_file)
            image = Image.open(image_path)
            row, col = divmod(i, num_cols)
            axes[row, col].imshow(image)
            axes[row, col].axis('off')
            axes[row, col].set_title(image_file, fontsize=8)
        
        # Hide unused subplots
        for i in range(len(image_files), num_rows * num_cols):
            row, col = divmod(i, num_cols)
            axes[row, col].axis('off')
        
        plt.tight_layout()
        plt.show()
```

```python
display_img= ImageProcessor()
display_img.display_images("runs/detect/train")
```
